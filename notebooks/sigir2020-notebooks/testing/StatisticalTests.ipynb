```python
import pandas as pd
from randtest import randtest
import numpy as np
```

```python
dfba = pd.read_csv("ba-scoresPerTopic.tsv", sep="\t")
dfct = pd.read_csv("ct-scoresPerTopic.tsv", sep="\t")
dfct.loc[:,"201801":"201940"]
```

```python
alpha = 0.05
def holmbonf(reslist):
    reslist.sort(key=lambda p : p[1])
    print(reslist)
    corrected = [alpha/(len(reslist)+1-k) for k in range(len(reslist))]
    print(corrected)
    thresh = -1
    for i in range(len(reslist)):
        if reslist[i][1] > corrected[i]:
            thresh = i
            break
    print(thresh)
```

```python
holmbonf([
  ('+Hypernyms', 0.00865),
 ('-Clause Weights', 0.00945),
 ('-Multi-word', 0.00185),
 ('-Positive', 0.04505),
 ('-Stopword Filtering', 0.0095)])
```

```python
bareslist = []
for i in range(len(dfba)):
    data = dfba.loc[i,"201701":"201940"]
    ref_data = dfba.loc[0,"201701":"201940"]
    result=randtest(data, ref_data, num_jobs=4, seed=0, alternative="less", num_permutations=20000)
    groupname = dfba.loc[i]["ablationgroup"]
    pval = result.p_value
    bareslist.append((groupname, pval))
    print("[ba]",groupname,pval)
bareslist
```

```python
ctreslist = []
for i in range(len(dfct)):
    data = dfct.loc[i,"201801":"201940"]
    ref_data = dfct.loc[0,"201801":"201940"]
    result=randtest(data, ref_data, num_jobs=4, seed=0, alternative="less", num_permutations=20000)
    groupname = dfct.loc[i]["ablationgroup"]
    pval = result.p_value
    ctreslist.append((groupname, pval))
    print("[ct]",groupname,pval)
ctreslist
```
